#### #1 상하좌우

**문제 설명**<br>
1) 지도 <br>
NXN 크기 정사각형 공간
1x1로 나누어져있음, (1,1)~(N,N)으로 표현

2) 이동계획서 <br>
U,D,L,R : 순서대로 상하좌우 한칸 이동<br>
공간 밖으로 나간다면 무시(RRRUDD일때 U는 무시)


```python
n = int(input())
plan = input().split()

x,y= 1,1

dx = [-1,1,0,0]
dy = [0,0,-1,1]
move_type = ['U','D','L','R']

for move in plan:
    for i in range(4):
        if move in move_type[i]:
            nx = x + dx[i]
            ny = y + dy[i]
    if nx < 1 or nx > n or ny < 1 or ny > n :
        continue
    
    x , y = nx, ny
    
print(x,y)
```

### #1 왕실의 나이트

**문제 설명**<br>

나이트가 정원에서 이동 할 수 있는 경우의 수를 구하여라.<br>
1. 정원<br>
정원은 8X8 평면이다. <br>
정원의 나이트 위치는 a1~h8으로 표현한다.<br>
2. 나이트<br>
나이트는 이동시 L자 형태로만 이동 가능(수평2,수직1/수직2,수평1) 하고 정원 밖으로 나갈 수 없다.<br>

```python
from string import ascii_lowercase

idx=input()
row=ascii_lowercase[:8]

x = row.find(idx[0])+1
y = int(idx[1:])

dx=[2,2,-2,-2,1,-1,1,-1]
dy=[1,-1,1,-1,2,2,-2,-2]

count = 0
for i in range(len(dx)):
    nx = x+dx[i]
    ny = y+dy[i]
    if nx<1 or nx>8 or ny<1 or ny>8:
        continue
    count+=1
print(count)
```

**내가 따로 추가한거** <br>

정원의 크기= n 으로 주어짐(nxn) <br>
정원 나이트의 위치에서 input 값이 a1=1a로 둘 다 input값으로 들어갈수 있음

```python
from string import ascii_lowercase

def solution(n,idx):
    row=ascii_lowercase[:n]

    try:
        x = row.find(idx[0])+1
        y = int(idx[1:])
    except:
        x = row.find(idx[-1])+1
        y = int(idx[:-1])

    dx=[2,2,-2,-2,1,-1,1,-1]
    dy=[1,-1,1,-1,2,2,-2,-2]

    count = 0
    for i in range(len(dx)):
        nx = x+dx[i]
        ny = y+dy[i]
        if nx<1 or nx>n or ny<1 or ny>n:
            continue
        count+=1
    print(count)
```

```python
n=8
idx='a1'
solution(n,idx)
```
